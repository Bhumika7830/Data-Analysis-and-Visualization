```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15.,10))
sns.countplot(x="STDepression",data=df,width=0.5)
```

```python
sns.histplot(df["Cholesterol"],bins=20,kde=True)
plt.title("cholesterol of patients ")
plt.xlabel("cholesterol")
plt.ylabel("count")
```

```python
sns.boxplot(data=df,x="ChestPainType",y="Cholesterol")
plt.title("Cholesterol level by Chestpain type")
```
